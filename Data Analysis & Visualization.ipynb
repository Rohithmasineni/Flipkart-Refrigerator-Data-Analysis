```python
# Box Plot to represent the Distribution of Refrigerator Capacities in Liters
plt.figure(figsize=(5,2))
sns.boxplot(x='Capacity_in_Liters', data = df4)
plt.title('Distribution of Refrigerator Capacities in Liters')
```

```python
# INTERPRETATION

The graph is box plot, illustrating the distribution of refrigerator capacities in liters:
'Median':
The line inside the box represents the median, which is around 185 liters
'Interquartile Range(IQR)':
The box itself spans the interquartile range, which is the range between the 25th percentile(Q1) and the 75th percentile(Q3).
This indicates that 50% of refrigerators have capacities falling within this range.
'Whiskers':
The lines extending from the box (whiskers) represent the data excluding outliers
'Outliers':
The outliers indicating a refrigerator with a capacity significantly higher than the rest
```

```python
# Hist Plot to represent Count of Refrigerators Having Different Type of Doors
plt.figure(figsize=(5,3))
plt.hist(x='Door_Type',data=df4, edgecolor='black')
plt.title('Count of Refrigerators Having Different Type of Doors')
plt.xlabel('Door Type')
plt.ylabel('Count of Refrigerators')
```

```python
# INTERPRETATION

--> The above graph is a Histogram which gives the information about the Count of Refrigerators Having Different Type of Doors
--> Here, we can see clearly that the majority of refrigerators are of Single Door type (i.e, around 420)
--> Double Door type (i.e, around 100) and Triple Door type (i.e, around 50)
```

```python
# Violin Plot to represent Distribution of Refrigerator Energy Efficiency Ratings
plt.figure(figsize=(5,3))
sns.violinplot(x='Energy_Efficiency_Rating', data = df4)
plt.title('Distribution of Refrigerator Energy Efficiency Ratings')
```

```python
# INTERPRETATION

--> The plot is a violin plot. It shows the distribution of Energy Efficiency Ratings for refrigerators.
--> The wider section of violin indicates the higher density of data points, while thinner sections of represent lower densities.
--> The white dot in the middle of violin plot represents the median value, and thick black line extends to the first and third quartiles
```

```python
# Box Plot for Distribution of Actual Prices
plt.figure(figsize=(5,3))
sns.boxplot(x='Actual_price',data=df4)
plt.title('Distribution of Actual Prices')
```

```python
# INTERPRETATION

--> The boxplot is visualizing the distribution of 'Actual_Price' data:
--> The box represents the IQR, which contains the middle 50% of the data. The left most of the box is the first quartile (25%), and the right most of the box is third quartile (75%)
--> The line inside the box represnts the median (50th percentile).
--> The whiskers extend from the box to the minimum and maximum values within 1.5 times of IQR
--> Any data points outside the whiskers are considered outliers but there are no outliers in this plot
```

```python
# Density Plot for Distribution of Offer Percentage
plt.figure(figsize=(5,3))
sns.kdeplot(data=df4['Offer_percentage'], fill=True)
plt.title('Distribution of Offer Percentage')
```

```python
# INTERPRETATION

--> the plot shown is a density plot depicting the distribution of offer percentages for different brands of refrigerators.
--> X-Axis: It shows the range of offer percentages provided on refrigerators, spanning from approximately 10% to 35%
--> Y-Axis: It represents the probability density of the offer percentages.
--> The curve indicates how the offer percentages are distributed across the brands.
--> There is a noticeable peak (mode) around the 25-30% offer range.
```

```python
# Violin Plot to represent Distribution of Selling Price
plt.figure(figsize=(5,3))
sns.violinplot(x='Selling_price',data=df4)
plt.title('Distribution of Selling Price')
```

```python
# INTERPRETATION

--> The plot is a violin plot.
--> The widest part of the violin represents the area where most data points lie, while the thinner parts indicate fewer data points.
--> The shape of this plot suggests a bimodal distribution which means there are two distinct peaks in the data.
--> The white dot inside the black thick line represents the median value of selling prices.
--> The thick black line in the violin shows the Interquartile range(IQR), which encompasses the middle 50% of data.
```

# Bi-Variate Analysis

This will help us in understanding the relationship between the variables.

```python
# Bar plot 
# Categorical vs Categorical variables

plt.figure(figsize=(8,4))
sns.countplot(x='Door_Type', hue='Brand_Name', data = df4)
plt.title('Brand-Wise Distribution of Refrigerator Door Types')
```

1) The count plot illustrates the distribution of different refrigerator door types across various brands. Here is an interpretation of the plot:
2) The x-axis represents different types of refrigerator doors (Single Door, Double Door, and Triple Door).
3) The y-axis shows the count of refrigerators with each door type.


Key observations:

1) Single Door refrigerators have the highest count across most brands. Godrej, Whirlpool, and LG show high counts in this category, with relatively    fewer units from brands like Haier and Candy.
2) Double Door refrigerators are notably more prominent in brands like Samsung, LG and Whirlpool, while Godrej have very lower counts, and there are no double door refrigerators in Haier and Candy.
3) Triple Door refrigerators have the lowest representation overall, with Whirlpool being the only significant contributor.

```python
# Categorical vs Categorical variables (crosstab)

cross_tab1 = pd.crosstab(df4['Brand_Name'],df4['Compressor_type'])
cross_tab1
```

```python
# Plotting the bar plot with adjusted figure size
cross_tab1.plot(kind='bar', stacked=True, figsize=(7,4))
plt.legend(title='Offer_percentage', fontsize='small', ncol=2)
plt.title('Stacked Bar Plot of Brand Name vs Compressor Type')
plt.xlabel('Brand Name')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(True)
```

1) The stacked bar plot depicts the distribution of various compressor types across different brands of appliances or products (as indicated by the   labels on the x-axis).
   Here is an interpretation of the plot:
2) The x-axis represents different brand names (Candy, Godrej, Haier, LG, Samsung, and Whirlpool).
3) The y-axis shows the count of instances for each compressor type within the brands.
4) The legend explains that different colors correspond to different compressor types (e.g., Digital Inverter Compressor, Fixed Speed Compressor, etc.).
5) Candy and Haier have a smaller count of compressors compared to other brands, with a prominent representation of the Normal Compressor.


```python
# Continuous vs Continuous 

colors = ['red', 'blue', 'green', 'yellow', 'purple', 'orange', 'pink', 'brown', 
          'gray', 'cyan', 'magenta', 'lime', 'teal', 'maroon', 'navy', 'olive', 
          'coral', 'gold']
plt.figure(figsize=(6,4))
sns.scatterplot(x='Actual_price', y='Selling_price', hue='Offer_percentage',palette=colors, data=df4)
plt.legend(title='Offer_percentage', fontsize='small', ncol=3)
plt.title('Impact of Offer Percentage on Selling Price')
plt.xticks(rotation=45)
plt.grid(True)
```

1) The scatter plot demonstrates the relationship between certain refrigerator's actual price, selling price, and the offer percentage.
2) As the actual price increases, the selling price tends to decrease. This suggests that higher offer percentages are applied to products with higher
   actual prices, resulting in a lower final selling price.

```python
# Continuous to Continuous variable (correlation plot)
df4[['Capacity_in_Liters','Selling_price','Offer_percentage']].corr()
```

```python
plt.figure(figsize=(6,3))
sns.heatmap(df4[['Capacity_in_Liters','Selling_price','Offer_percentage']].corr(),cmap='coolwarm' ,annot=True)
plt.title('Correlation Matrix for Product Attributes')
plt.show
```

1) The heatmap displays the correlation between three variables:                                                                                       
    a) Capacity in Liters                                                                                                                               
    b) Selling Price                                                                                                                                    
    c) Offer Percentage

2) There is a strong positive correlation (0.93) between Capacity in Liters and Selling Price, indicating that as the capacity increases,
    the selling price tends to increase as well.
3) A moderate positive correlation exists between Capacity in Liters and Offer Percentage (0.39), suggesting that higher capacity products might
    have slightly higher offer percentages                                                                                                              
4) Similarly, there is a moderate positive correlation (0.37) between Selling Price and Offer Percentage, indicating discounts or offers maybe less.

```python
# Continuous vs Categorical variables
cross_tab = pd.crosstab(df4['Brand_Name'],df4['Rating'])
cross_tab
```

```python
cross_tab.plot(kind='bar', figsize=(6, 3), width=0.8)
plt.legend(title='Rating', fontsize='small', ncol=2)
plt.title('Brand Rating Comparison')
plt.xticks(rotation=0)
plt.grid(True)
```

1) The bar plot compares brand ratings for different brands. Each bar represents a rating category.
2) Whirlpool and LG appear to have higher bars, indicating more products with higher ratings, whereas Candy and Haier have fewer or lower
   ratings overall.
3) This comparision highlights the varying performance and customer ratings across different brands.

```python
# Continuous vs Categorical
plt.figure(figsize=(8,3))
sns.scatterplot(y='Energy_Efficiency_Rating', x='Selling_price', hue='Brand_Name', data=df4)
plt.legend(title='Brand_Name', fontsize='small', ncol=1)
plt.title('Energy Efficiency Ratings Across Brands with Price Comparision')
```

INTERPRETATION:   

--> The scatter plot shows the 'Energy Efficiency Ratings Across Brands with Price Comparision'. It compares different brands based on their Energy Efficiency Ratings and their selling price.                                                                                                               
1) We can see that most of the data points are falling between ratings of 2 and 5.
2) Brands such as LG and Whirlpool tend to have relatively higher energy efficiency ratings and are spread across various price ranges.
3) Godrej and Haier models being clustered in certain specific price ranges, instead of being spread out across all price ranges.
   This indicates that these brands might focus on offering produts within a particular price bracket to target a specific group of customers, like budjet friendly or mid-range buyers.
4) Some brands, such as Candy have lower energy efficiency ratings, which are generally priced lower.

5) The scatter plot suggests that energy efficiency rating and price may have some correlation, where higher ratings might lead to a higher price,
   though this varies across the brands
