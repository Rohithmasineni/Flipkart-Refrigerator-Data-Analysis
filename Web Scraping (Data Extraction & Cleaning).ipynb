```python
#--> Extracting the data from each page 
#--> Initialized empty lists to append the extracted data into them and to create the DataFrame

# Empty Lists
Product_Name = []
Actual_Price = []
Offer_Percentage = []
Selling_Price = []
Rating = []
Features = []

# request header to get data from each page
request_header = {'Content-Type':'https://www.flipkart.com/search?q=refrigerators&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'}
# Count of Page Numbers
PageNum= []
# Looping over each page
for i in range(1,25):
    start = time.time()
    URL = 'https://www.flipkart.com/search?q=refrigerators&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page = {}'.format(i)
    page = requests.get(URL,headers = request_header)
    pagecontent = page.text
    soup = BeautifulSoup(pagecontent)
    
    # Initialized inner loop to iterate over each piece of data
    for x in soup.find_all('div',attrs={'class':'yKfJKb row'}):
        product_name = x.find('div',attrs={'class':'KzDlHZ'})
        actual_price = x.find('div',attrs={'class':'yRaY8j ZYYwLA'})
        offer_percentage = x.find('div',attrs={'class':'UkUFwK'})
        selling_price = x.find('div',attrs={'class':'Nx9bqj _4b5DiR'})
        ratings = x.find('span',attrs={'class':'Y1HWO0'})
        features = x.find('ul',attrs={'class':'G4BRas'})

        # Conditional Statements to append the data into the empty lists
        if product_name is None:
            Product_Name.append(np.NaN)
        else:
            Product_Name.append(product_name.text)
    
        if actual_price is None:
            Actual_Price.append(np.NaN)
        else:
            Actual_Price.append(actual_price.text)
    
        if offer_percentage is None:
            Offer_Percentage.append(np.NaN)
        else:
            Offer_Percentage.append(offer_percentage.text)
    
        if selling_price is None:
            Selling_Price.append(np.NaN)
        else:
            Selling_Price.append(selling_price.text)
    
        if ratings is None:
            Rating.append(np.NaN)
        else:
            Rating.append(ratings.text)

        if features is None:
            Features.append(np.NaN)
        else:
            Features.append(features.text)
    
    #Time taken to get the data from each page
    end=time.time()
    print('Page {} completed in {} seconds'.format(i,end-start))
```

# Data Manipulation
Extracting new columns from the raw dataset

```python
# Extracting capacity of refrigerator from productname
pattern = r'\b(\d{3})\b'
df['Capacity_in_Liters'] = [re.search(pattern, str(row)).group(0) if re.search(pattern, str(row)) else None for row in df['Product_Name']]
```

```python
# Extracting the no.of doors that each refrigerator is having

pattern = r'\b(Single Door|Double Door|Triple Door)\b'
# Extract the matching text or assign NaN if not found
df['Door_Type'] = [re.search(pattern, str(row)).group(0) if re.search(pattern, str(row)) else np.nan for row in df['Product_Name']]
```

```python
# Extracting name of the brand of each refrigerator

pattern = r'^\w+'

# Extract the brand name using regex
df['Brand_Name'] = [re.search(pattern, str(row)).group(0) if re.search(pattern, str(row)) else None for row in df['Product_Name']]
```

```python
# Extracting the Energy Efficiency Rating of each refrigerator

pattern = r'\b(\d{1})\b'
df['Energy_Efficiency_Rating'] = [re.search(pattern, str(row)).group(0) if re.search(pattern, str(row)) else None for row in df['Product_Name']]
#Refrigerators_Data['Energy_Efficiency_Rating'] = Refrigerators_Data['Energy_Efficiency_Rating'].astype(int)

# re.search(pattern, str(row)) finds the match in each row.
# .group(0) extracts the matched string if found.
```

```python
# Removing special characters from Actual_Price column

df['Actual_price'] = [re.sub(r'[^\d]', '', str(row)) for row in df['Actual_Price']]
# re.sub(r'[^\d]', '', str(row)) removes all characters except digits (\d)

# Convert the extracted numeric string to an integer (optional)
df['Actual_price'] = df['Actual_price'].astype(int)
```

```python
# Removing special characters from Offer_Percentage column

df['Offer_percentage'] = [re.sub(r'[^\d]', '',str(row)) for row in df['Offer_Percentage']]

# Convert the extracted numeric string to an integer (optional)
df['Offer_percentage'] = df['Offer_percentage'].astype(int)
```

```python
# Removing special characters from Selling_Price column

df['Selling_price'] = [re.sub(r'[^\d]', '',str(row)) for row in df['Selling_Price']]

# Convert the extracted numeric string to an integer (optional)
df['Selling_price'] = df['Selling_price'].astype(int)
```

```python
# Extracting type of compressor from features column

pattern = r'^(Inverter Compressor|Reciprocating Compressor|Normal Compressor|Digital Inverter Compressor|Fixed Speed Compressor|Smart Inverter Compressor)'
df['Compressor_type'] = [' ,'.join(set(re.findall(pattern, str(row)))) if re.findall(pattern, str(row)) else None for row in df['Features']]
```
